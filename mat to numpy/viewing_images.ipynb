### Displaying Images

```python
for i in range(1,767,15):
    f = h5py.File(str(i)+'.mat','r')
    gr = f.get('cjdata')
    im = np.array(gr.get('image'))
    plt.imshow(im, cmap="gray") 
    plt.show()
    print((gr.get('label')[()])) #cjdata.label: 1 for meningioma, 2 for glioma, 3 for pituitary tumor
```
